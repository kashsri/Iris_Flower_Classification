```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/IRIS.csv')
df.head(  )
```

```python
df['species'],categories = pd.factorize(df['species'])
df.head()
```

```python
df.isna().sum()
```

```python
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species");
```

```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species");
```

```python
k_rng = range(1,10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
plt.xlabel('k_rng')
plt.ylabel("Sum of Squared errors")
plt.plot(k_rng, sse)
```

```python
km = KMeans(n_clusters=3,random_state=0,)
y_predicted = km.fit_predict(df[['petal_length', 'petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(150)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
true_lables = df.species
predicted_lables = df.cluster

cm = confusion_matrix(true_lables, predicted_lables)
class_lables = ['Setosa', 'versicolor', 'verginica']

# plot confusion matrics
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion_Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_lables))
plt.xticks(tick_marks, class_lables)
plt.yticks(tick_marks, class_lables)

# Fill matrix with values
for i in range(len(class_lables)):
  for j in range(len(class_lables)):
    plt.text(j, i , str(cm[i][j]), ha='center', va='center', color='white')

plt.xlabel('predicted label')
plt.ylabel('True label')
plt.show()
```
